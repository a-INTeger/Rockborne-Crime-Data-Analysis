```python
# import the libraries we need for preprocessing
from glob import glob
import pandas as pd
import numpy as np
import datetime
```

Import all the data from all the regions using `glob`

```python
# all the regions of interest
REGIONS = ["merseyside", "metropolitan", "thames-valley", "west-midlands"]

all_dfs = dict()

for region in REGIONS:
    # glob simply returns all the relative paths matching a certain regex
    files = glob(f"./Data/Raw/*/*{region}-street.csv")

    region_dfs = list()
    for file in files:
        # reach each csv across all sub-directories
        df = pd.read_csv(file)
        region_dfs.append(df)

    # concatenate the subdirectory dataframes into one larger region dataframe
    final_df = pd.concat(region_dfs)
    all_dfs[region] = final_df

for region in REGIONS:
    display(all_dfs[region].head(3))
```

```python
# understanding the types of the data
for region in REGIONS:
    print(all_dfs[region].info())
    print("-"*64)
```

The 'Month' columns is of the incorrect type `str` instead of `Date`

```python
# convert the month column to datetime
for region in REGIONS:
    all_dfs[region]["Month"] = all_dfs[region]["Month"].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m"))
    all_dfs[region].info()
```

Identifying any nulls and things to fix

```python
countNull = pd.DataFrame(columns=all_dfs["merseyside"].columns) # Null count
perNull = pd.DataFrame(columns=all_dfs["merseyside"].columns)  # percentage of values that are null

for region in REGIONS:
    countNull.loc[len(countNull)] = dict(all_dfs[region].isnull().sum())
    perNull.loc[len(perNull)] = dict(all_dfs[region].isnull().mean())
```

```python
countNull, perNull = countNull.set_index(pd.Index(REGIONS)), perNull.set_index(pd.Index(REGIONS)) # reindex for readability
display(countNull, perNull)
```

# Data Dictionary

| Column name     | Column Description     |
| ------------- | ------------- |
| Crime ID | Unique Identifier for each crime, is null for minor infractions|
| Month | year and month when the crime occured|
| Reported By | Police force that reported the crime |

| Latitude | Latitude where the crime occured |

| LSOA code | Lower layer Super Output Area (LSOA) code code - smaller areas for further population analysis |
| LSOA name | Name corresponding to the LSOA code|
| Crime type | Type of offense that was committed|
| Last outcome category | Outcome of the event |
| Context | Any additional context surrounding the event |


## Updates
- Removed the Context column (all blanks)
- Removed the Location descriptor (Not necessary due to latitude and logitude provided)
- Dropped all values with unknown Latitude and Longitude (Cannot generate maps during EDA later)

```python
# dropping unneeded collumns / null values
for region in REGIONS:
    all_dfs[region] = all_dfs[region].drop(columns=["Context", "Location"])
    all_dfs[region] = all_dfs[region].dropna(subset=["Latitude", "Longitude"])
    print(all_dfs[region]["Month"].value_counts())
    display(all_dfs[region])
```

# Importing Other data (police force population data)

```python
# read csv
populationData = pd.read_csv("./Data/Raw/policeforcepopdata.csv")
# extract recent data
relevantPopData = populationData[(populationData["Year"] == 2024) & (populationData["PFA 2023 Name"].isin(["Metropolitan Police", "Thames Valley", "Merseyside", "West Midlands"]))]
# set index properly
relevantPopData = relevantPopData.set_index("PFA 2023 Name")
# drop unneeded columns
relevantPopData = relevantPopData.drop(columns=["PFA 2023 Code", "Year"])
# rename index columns rpoper;y
relevantPopData = relevantPopData.set_index(pd.Index(["metropolitan", "merseyside", "west-midlands", "thames-valley"]))

# convert each column to a proper int
for column in relevantPopData.columns:
    relevantPopData[column] = relevantPopData[column].str.replace(",", "")

# add a total columns
relevantPopData["Totals"] = relevantPopData.astype("int64").sum(axis = 1 )
display(relevantPopData)
```

Grouping Age population data for ease of use

```python
popDataDfs = list() # list of all dfs

# iterate over all regions
for region in REGIONS: 
    # melt dataset to get everything into 2 columns
    long = relevantPopData.loc[[region], :].melt(var_name="GenderAge", value_name="Value")

    # list of each age
    x = range(86)
    # result list
    a = list()

    # bins to split into
    bins = [0, 18, 35, 50, 65, 81, 100]

    # merging between genders for each age
    for i in range(86):
        a.append(long[(long["GenderAge"] == f"F{i}") | (long["GenderAge"] == f"M{i}")]["Value"].astype(int).sum())

    # new dataframe with age and population
    ageData = pd.DataFrame()
    ageData["Age"] = x
    ageData["Pop"] = a

    # assign each age to an age bin
    ageData["age_bin"] = pd.cut(ageData["Age"], bins=bins, labels=["0-17", "18-34", "35-49", "50-64", "65-80", "81+"], right=False)
    # group by the bins and sum them
    result = pd.DataFrame(ageData.groupby("age_bin")["Pop"].sum()).T
    popDataDfs.append(result)

# concatonate results and save
ageData = pd.concat(popDataDfs, ignore_index=True)
ageData = ageData.set_index(pd.Index(REGIONS))
ageData
ageData.to_csv("./Data/Clean/binnedAgeGroups.csv")
```

```python
# concatonate all dfs to one single dfs to save
FINAL = pd.concat([all_dfs[region] for region in REGIONS])
FINAL.to_csv("./Data/Clean/crime-data.csv")
```

### Importing Outcome data (Identical process to crime data)

```python
REGIONS = ["merseyside", "metropolitan", "thames-valley", "west-midlands"]

outcomes = dict()

for region in REGIONS:
    files = glob(f"./Data/Raw/*/*{region}-outcomes.csv")

    region_dfs = list()
    for file in files:
        df = pd.read_csv(file)
        region_dfs.append(df)

    final_df = pd.concat(region_dfs)
    outcomes[region] = final_df

for region in REGIONS:
    display(outcomes[region].head(3))
```

```python
for region in REGIONS:
    print(outcomes[region].info())
```

```python
outcomeNull = pd.DataFrame(columns=outcomes["merseyside"].columns)
outcomeNullPer = pd.DataFrame(columns=outcomes["merseyside"].columns)
for region in REGIONS:
    outcomeNull.loc[len(outcomeNull)] = dict(outcomes[region].isnull().sum())
    outcomeNullPer.loc[len(outcomeNullPer)] = dict(outcomes[region].isnull().mean())

outcomeNull, outcomeNullPer = outcomeNull.set_index(pd.Index(REGIONS)), outcomeNullPer.set_index(pd.Index(REGIONS))
display(outcomeNull, outcomeNullPer)
```

A small minority of values have a blank longitude, latitude or LSOA. Therefore we can drop these values.

```python
for region in REGIONS:
    outcomes[region] = outcomes[region].drop(columns=["Location"])
    outcomes[region] = outcomes[region].dropna(subset=["Longitude", "Latitude", "LSOA code", "LSOA name"])
```

```python
OUTCOMES = pd.concat([outcomes[region] for region in REGIONS])
OUTCOMES.to_csv("./Data/Clean/outcomes.csv")
```
