```python
from glob import glob
import pandas as pd
import numpy as np
```

```python
REGIONS = ["merseyside", "metropolitan", "thames-valley", "west-midlands"]

all_dfs = list()

for region in REGIONS:
    files = glob(f"./*/*/*{region}-street.csv")

    dfs = list()
    for file in files:
        df = pd.read_csv(file)
        dfs.append(df)

    final_df = pd.concat(dfs)
    all_dfs.append(final_df)

display(all_dfs[0].head(), all_dfs[1].head(), all_dfs[2].head(), all_dfs[3].head())
```
